```python
import urllib, urllib.request
from bs4 import BeautifulSoup
import pandas as pd

queries = [
    'Stocks',
    'Bonds',
    'Investments',
    'Portfolio',
    'Risk+management',
    'Financial+markets',
    'Asset+allocation',
    'Capital+budgeting',
    'Financial+planning',
    'Valuation',
    'Derivatives',
    'Hedge+funds',
    'Cash+flow',
    'Asset+pricing',
    'Financial+statements',
    'Mutual+funds',
    'Risk+assessment',
    'Financial+modeling',
    'Market+analysis',
    'Financial+derivatives'
 ]

n = 200
urls = [f'http://export.arxiv.org/api/query?search_query=all:{query}&start=0&max_results={n}' for query in queries]
data = [urllib.request.urlopen(url) for url in urls]
content = [elem.read().decode('utf-8') for elem in data]
print(content[-1])
```

```python
# из полученных данных нужны только аннотации (summary) и тематические категории (category)
# некоторым статьям присваивается несколько категорий, поэтому поиск ведется только по arxiv:primary_category

data = {"summary": [], "category": []}
df = pd.DataFrame(data)

for elem in content:
  parsed = BeautifulSoup(elem, "html.parser")
  summary = [x.text for x in parsed.find_all("summary")]
  category = [x["term"] for x in parsed.find_all("arxiv:primary_category")]
  for x in range(len(summary)):
    df.loc[len(df)] = [summary[x], category[x]]
```
